# Breadth First Search

```python
class Graph:
    def __init__(self):
        self.adjlist = {}

    def addedge(self, u, v):
        if u not in self.adjlist:
            self.adjlist[u] = []
        if v not in self.adjlist:
            self.adjlist[v] = []
        self.adjlist[u].append(v)
        self.adjlist[v].append(u)

    def bfs(self, start):
        visited = set()
        queue = [start]
        visited.add(start)

        while queue:
            node = queue.pop(0)
            print(node, end=' ')
            for neighbor in self.adjlist.get(node, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)

graph = Graph()
graph.addedge(1, 2)
graph.addedge(1, 3)
graph.addedge(2, 4)
graph.addedge(2, 5)
graph.addedge(3, 6)
graph.addedge(3, 7)
print("BFS:")
graph.bfs(1)
```

# Depth First Search

```python
class Graph:
    def __init__(self):
        self.adj_list = {}

    def addedge(self, u, v):
        if u not in self.adj_list:
            self.adj_list[u] = []
        if v not in self.adj_list:
            self.adj_list[v] = []
        self.adj_list[u].append(v)
        self.adj_list[v].append(u)

    def dfs(self, start):
        visited = set()
        stack = [start]

        while stack:
            node = stack.pop()
            if node not in visited:
                print(node, end=' ')
                visited.add(node)
                for neighbor in reversed(self.adj_list.get(node, [])):
                    if neighbor not in visited:
                        stack.append(neighbor)

graph = Graph()
graph.addedge(1, 2)
graph.addedge(1, 3)
graph.addedge(2, 4)
graph.addedge(2, 5)
graph.addedge(3, 6)
graph.addedge(3, 7)
print("DFS:")
graph.dfs(1)
```

# Uniform Cost Search

```python
class PriorityQueue:
    def __init__(self):
        self.elements = []

    def isEmpty(self):
        return len(self.elements) == 0

    def put(self, item, priority):
        self.elements.append((priority, item))
        self.elements.sort(reverse=True) 

    def get(self):
        return self.elements.pop()

class Graph:
    def __init__(self):
        self.adjlist = {}

    def addEdge(self, u, v, weight):
        if u not in self.adjlist:
            self.adjlist[u] = []
        if v not in self.adjlist:
            self.adjlist[v] = []
        self.adjlist[u].append((v, weight))
        self.adjlist[v].append((u, weight))

    def uniformCostSearch(self, start, goal):
        pq = PriorityQueue()
        pq.put(start, 0)
        visited = set()
        costToNode = {start: 0}
        parent = {start: None}

        while not pq.isEmpty():
            currentCost, node = pq.get()
            if node in visited:
                continue
            visited.add(node)

            if node == goal:
                self.printPath(parent, goal)
                return currentCost

            for neighbor, weight in self.adjlist.get(node, []):
                if neighbor in visited:
                    continue
                newCost = currentCost + weight
                if neighbor not in costToNode or newCost < costToNode[neighbor]:
                    costToNode[neighbor] = newCost
                    parent[neighbor] = node
                    pq.put(neighbor, newCost)

        print("no path found")
        return float('inf')

    def printPath(self, parent, goal):
        path = []
        while goal is not None:
            path.append(str(goal))
            goal = parent[goal]
        print("Path:", ' -> '.join(reversed(path)))


g = Graph()
g.addEdge(1, 2, 1)
g.addEdge(1, 3, 4)
g.addEdge(2, 3, 2)
g.addEdge(2, 4, 5)
g.addEdge(3, 4, 1)
print("Uniform Cost Search")
cost = g.uniformCostSearch(1, 4)
print("Cost:", cost)
```

# Bidirectional Search

```python
class Graph:
    def __init__(self):
        self.adjList = {}

    def addEdge(self, u, v):
        if u not in self.adjList:
            self.adjList[u] = []
        if v not in self.adjList:
            self.adjList[v] = []
        self.adjList[u].append(v)
        self.adjList[v].append(u)

    def bidirectionalSearch(self, start, goal):
        if start == goal:
            return [start]

        forwardQueue = [start]
        backwardQueue = [goal]
        forwardVisited = {start: None}
        backwardVisited = {goal: None}

        while forwardQueue and backwardQueue:
            forwardPath = self._expandLayer(forwardQueue, forwardVisited, backwardVisited)
            if forwardPath:
                return forwardPath

            backwardPath = self._expandLayer(backwardQueue, backwardVisited, forwardVisited)
            if backwardPath:
                return backwardPath

        print("No path found")
        return []

    def _expandLayer(self, queue, visitedFromThisSide, visitedFromOtherSide):
        current = queue.pop(0)
        for neighbor in self.adjList.get(current, []):
            if neighbor not in visitedFromThisSide:
                queue.append(neighbor)
                visitedFromThisSide[neighbor] = current
                if neighbor in visitedFromOtherSide:
                    return self._constructPath(neighbor, visitedFromThisSide, visitedFromOtherSide)
        return None

    def _constructPath(self, meetingPoint, forwardVisited, backwardVisited):
        path = []

        node = meetingPoint
        while node is not None:
            path.append(node)
            node = forwardVisited[node]
        path = path[::-1]

        node = backwardVisited[meetingPoint]
        while node is not None:
            path.append(node)
            node = backwardVisited[node]

        return path

g = Graph()
g.addEdge(1, 2)
g.addEdge(1, 3)
g.addEdge(2, 4)
g.addEdge(3, 5)
g.addEdge(4, 6)
g.addEdge(5, 6)
print("Bidirectional Search:")
path = g.bidirectionalSearch(1, 6)
print("Path:", path)
```

# Best First Search

```python
class Graph:
    def __init__(self):
        self.adjList = {}

    def addEdge(self, u, v, weight):
        if u not in self.adjList:
            self.adjList[u] = []
        if v not in self.adjList:
            self.adjList[v] = []
        self.adjList[u].append((v, weight))
        self.adjList[v].append((u, weight))

    def bestFirstSearch(self, start, goal, heuristic):
        openList = [(start, heuristic(start, goal))]
        visited = set()
        parent = {start: None}

        while openList:
            openList.sort(key=lambda x: x[1])
            node, _ = openList.pop(0)
            if node in visited:
                continue
            visited.add(node)

            if node == goal:
                self.printPath(parent, goal)
                return

            for neighbor, _ in self.adjList.get(node, []):
                if neighbor not in visited:
                    estimatedCost = heuristic(neighbor, goal)
                    openList.append((neighbor, estimatedCost))
                    parent[neighbor] = node

        print("No path found")

    def printPath(self, parent, goal):
        path = []
        while goal is not None:
            path.append(str(goal))
            goal = parent[goal]
        print("Path:", ' -> '.join(reversed(path)))

def heuristic(node, goal):
    return abs(node - goal)

g = Graph()
g.addEdge(1, 2, 1)
g.addEdge(1, 3, 4)
g.addEdge(2, 4, 2)
g.addEdge(3, 5, 1)
g.addEdge(4, 6, 5)
g.addEdge(5, 6, 2)
print("Best First Search:")
g.bestFirstSearch(1, 6, heuristic)
```

# A* Search

```python
class Graph:
    def __init__(self):
        self.adjList = {}

    def addEdge(self, u, v, weight):
        if u not in self.adjList:
            self.adjList[u] = []
        if v not in self.adjList:
            self.adjList[v] = []
        self.adjList[u].append((v, weight))
        self.adjList[v].append((u, weight))

    def aStarSearch(self, start, goal, heuristic):
        openList = [(0, start)]
        visited = set()
        costToNode = {start: 0}
        parent = {start: None}

        while openList:
            openList.sort() 
            currentCost, node = openList.pop(0)
            if node in visited:
                continue
            visited.add(node)

            if node == goal:
                self.printPath(parent, goal)
                return currentCost

            for neighbor, weight in self.adjList.get(node, []):
                if neighbor in visited:
                    continue
                newCost = costToNode[node] + weight
                estimatedCost = newCost + heuristic(neighbor, goal)
                if neighbor not in costToNode or newCost < costToNode[neighbor]:
                    costToNode[neighbor] = newCost
                    parent[neighbor] = node
                    openList.append((estimatedCost, neighbor))

        print("No path found")
        return float('inf')

    def printPath(self, parent, goal):
        path = []
        while goal is not None:
            path.append(str(goal))
            goal = parent[goal]
        print("Path:", ' -> '.join(reversed(path)))

def heuristic(node, goal):
    return abs(node - goal)

g = Graph()
g.addEdge(1, 2, 1)
g.addEdge(1, 3, 4)
g.addEdge(2, 4, 2)
g.addEdge(3, 5, 1)
g.addEdge(4, 6, 5)
g.addEdge(5, 6, 2)
print("A* Search:")
cost = g.aStarSearch(1, 6, heuristic)
print("Cost:", cost)
```

# Hill Climbing Search

```python
class Graph:
    def __init__(self):
        self.adjList = {}

    def addEdge(self, u, v, weight):
        if u not in self.adjList:
            self.adjList[u] = []
        if v not in self.adjList:
            self.adjList[v] = []
        self.adjList[u].append((v, weight))
        self.adjList[v].append((u, weight))

    def hillClimbingSearch(self, start, goal, heuristic):
        current = start
        while current != goal:
            neighbors = self.adjList.get(current, [])
            if not neighbors:
                print("No path found")
                return float('inf')
            nextNode = min(neighbors, key=lambda x: heuristic(x[0], goal))
            if heuristic(nextNode[0], goal) >= heuristic(current, goal):
                print("No better neighbors found")
                return float('inf')
            current = nextNode[0]
            print(current, end=' ')
        print()
        return heuristic(start, goal) - heuristic(current, goal)

def heuristic(node, goal):
    return abs(node - goal)

g = Graph()
g.addEdge(1, 2, 1)
g.addEdge(1, 3, 4)
g.addEdge(2, 4, 2)
g.addEdge(3, 5, 1)
g.addEdge(4, 6, 5)
g.addEdge(5, 6, 2)
print("Hill Climbing Search:")
cost = g.hillClimbingSearch(1, 6, heuristic)
print("Cost:", cost)
```

# Simulated Anealing

```python
import math
import random

class Graph:
    def __init__(self):
        self.adjList = {}

    def addEdge(self, u, v, weight):
        if u not in self.adjList:
            self.adjList[u] = []
        if v not in self.adjList:
            self.adjList[v] = []
        self.adjList[u].append((v, weight))
        self.adjList[v].append((u, weight))

    def simulatedAnnealing(self, start, goal, heuristic, initialTemp, coolingRate):
        current = start
        currentTemp = initialTemp
        while currentTemp > 1:
            neighbors = self.adjList.get(current, [])
            if not neighbors:
                print("No path found")
                return float('inf')
            nextNode = random.choice(neighbors)[0]
            costDiff = heuristic(nextNode, goal) - heuristic(current, goal)
            if costDiff < 0 or random.uniform(0, 1) < math.exp(-costDiff / currentTemp):
                current = nextNode
            currentTemp *= coolingRate
            print(current, end=' ')
            if current == goal:
                break
        print()
        return heuristic(start, goal) - heuristic(current, goal)

def heuristic(node, goal):
    return abs(node - goal)

g = Graph()
g.addEdge(1, 2, 1)
g.addEdge(1, 3, 4)
g.addEdge(2, 4, 2)
g.addEdge(3, 5, 1)
g.addEdge(4, 6, 5)
g.addEdge(5, 6, 2)
print("Simulated Annealing Search:")
cost = g.simulatedAnnealing(1, 6, heuristic, 100, 0.99)
print("Cost:", cost)
```

# Local Beam Search

```python
class Graph:
    def __init__(self):
        self.adjList = {}

    def addEdge(self, u, v, weight):
        if u not in self.adjList:
            self.adjList[u] = []
        if v not in self.adjList:
            self.adjList[v] = []
        self.adjList[u].append((v, weight))
        self.adjList[v].append((u, weight))

    def localBeam(self, start, goal, heuristic, beamWidth):
        pq = [(heuristic(start, goal), start)]
        visited = set()
        while pq:
            nextPq = []
            for _ in range(len(pq)):
                _, node = pq.pop(0)
                if node in visited:
                    continue
                visited.add(node)
                if node == goal:
                    print("Goal found:", node)
                    return
                for neighbor, _ in self.adjList.get(node, []):
                    if neighbor not in visited:
                        nextPq.append((heuristic(neighbor, goal), neighbor))
            pq = sorted(nextPq)[:beamWidth]
        print("No path found")

def heuristic(node, goal):
    return abs(node - goal)

g = Graph()
g.addEdge(1, 2, 1)
g.addEdge(1, 3, 4)
g.addEdge(2, 4, 2)
g.addEdge(3, 5, 1)
g.addEdge(4, 6, 5)
g.addEdge(5, 6, 2)
print("Local Beam Search from 1 to 6:")
g.localBeam(1, 6, heuristic, 2)
```

# Minimax

```python
class Game:
    def __init__(self, board):
        self.board = board

    def isMovesLeft(self):
        return any(' ' in row for row in self.board)

    def evaluate(self):
        for row in self.board:
            if row[0] == row[1] == row[2] != ' ':
                return 10 if row[0] == 'X' else -10
        for col in range(3):
            if self.board[0][col] == self.board[1][col] == self.board[2][col] != ' ':
                return 10 if self.board[0][col] == 'X' else -10
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != ' ':
            return 10 if self.board[0][0] == 'X' else -10
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != ' ':
            return 10 if self.board[0][2] == 'X' else -10
        return 0

    def minimax(self, depth, isMax):
        score = self.evaluate()
        if score == 10:
            return score - depth
        if score == -10:
            return score + depth
        if not self.isMovesLeft():
            return 0
        if isMax:
            best = -float('inf')
            for i in range(3):
                for j in range(3):
                    if self.board[i][j] == ' ':
                        self.board[i][j] = 'X'
                        best = max(best, self.minimax(depth + 1, not isMax))
                        self.board[i][j] = ' '
            return best
        else:
            best = float('inf')
            for i in range(3):
                for j in range(3):
                    if self.board[i][j] == ' ':
                        self.board[i][j] = 'O'
                        best = min(best, self.minimax(depth + 1, not isMax))
                        self.board[i][j] = ' '
            return best

    def findBestMove(self):
        bestVal = -float('inf')
        bestMove = (-1, -1)
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == ' ':
                    self.board[i][j] = 'X'
                    moveVal = self.minimax(0, False)
                    self.board[i][j] = ' '
                    if moveVal > bestVal:
                        bestMove = (i, j)
                        bestVal = moveVal
        return bestMove

board = [
    ['X', 'O', 'X'],
    ['O', 'X', ' '],
    [' ', ' ', 'O']
]

game = Game(board)
bestMove = game.findBestMove()
print("The best move is:", bestMove)
```

# CSP

```python
class CSP:
    def __init__(self, mapGraph, colors):
        self.mapGraph = mapGraph
        self.colors = colors
        self.colorsAssigned = {}

    def isSafe(self, node, color):
        for neighbor in self.mapGraph.get(node, []):
            if self.colorsAssigned.get(neighbor) == color:
                return False
        return True

    def backtrack(self, node):
        if node is None:
            return True
        for color in self.colors:
            if self.isSafe(node, color):
                self.colorsAssigned[node] = color
                nextNode = self.nextNode(node)
                if self.backtrack(nextNode):
                    return True
                del self.colorsAssigned[node]
        return False

    def nextNode(self, node):
        nodes = list(self.mapGraph.keys())
        if node in nodes:
            index = nodes.index(node)
            if index + 1 < len(nodes):
                return nodes[index + 1]
        return None

    def solve(self):
        startNode = list(self.mapGraph.keys())[0] if self.mapGraph else None
        if self.backtrack(startNode):
            self.print()
        else:
            print("No solution exists")

    def print(self):
        for node, color in self.colorsAssigned.items():
            print(f"Region {node} -> Color {color}")
        print()

mapGraph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C']
}
colors = ['Red', 'Green', 'Blue']
csp = CSP(mapGraph, colors)
csp.solve()
```

# CSP With Forward Checking

```python
class CSPForward:
    def __init__(self, mapGraph, colors):
        self.mapGraph = mapGraph
        self.colors = colors
        self.colorsAssigned = {}
        self.domains = {node: set(colors) for node in mapGraph}

    def isSafe(self, node, color):
        for neighbor in self.mapGraph.get(node, []):
            if self.colorsAssigned.get(neighbor) == color:
                return False
        return True

    def forwardCheck(self, node, color):
        for neighbor in self.mapGraph.get(node, []):
            if neighbor not in self.colorsAssigned:
                if color in self.domains[neighbor]:
                    self.domains[neighbor].remove(color)
                    if not self.domains[neighbor]:
                        return False
        return True

    def restoreDomains(self, node, color):
        for neighbor in self.mapGraph.get(node, []):
            if neighbor not in self.colorsAssigned and color not in self.domains[neighbor]:
                self.domains[neighbor].add(color)

    def backtrack(self):
        if len(self.colorsAssigned) == len(self.mapGraph):
            return True
        node = self.unsassign()
        for color in self.domains[node]:
            if self.isSafe(node, color):
                self.colorsAssigned[node] = color
                originalDomains = {n: set(d) for n, d in self.domains.items()}
                if self.forwardCheck(node, color):
                    if self.backtrack():
                        return True
                self.colorsAssigned.pop(node)
                self.restoreDomains(node, color)
                self.domains = originalDomains
        return False

    def unsassign(self):
        for node in self.mapGraph:
            if node not in self.colorsAssigned:
                return node
        return None

    def solve(self):
        if self.backtrack():
            self.print()
        else:
            print("No solution exists")

    def print(self):
        for node, color in self.colorsAssigned.items():
            print(f"Region {node} -> Color {color}")
        print()

mapGraph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C']
}
colors = ['Red', 'Green', 'Blue']
csp = CSPForward(mapGraph, colors)
csp.solve()
```

# CSP With AC3

```python
class CSPAC3:
    def __init__(self, mapGraph, colors):
        self.mapGraph = mapGraph
        self.colors = colors
        self.colorsAssign = {}
        self.domains = {node: set(colors) for node in mapGraph}

    def isConsistent(self, node, color):
        for neighbor in self.mapGraph.get(node, []):
            if neighbor in self.colorsAssign and self.colorsAssign[neighbor] == color:
                return False
        return True

    def ac3(self):
        queue = [(xi, xj) for xi in self.mapGraph for xj in self.mapGraph[xi]]
        while queue:
            xi, xj = queue.pop(0)
            if self.revise(xi, xj):
                if not self.domains[xi]:
                    return False
                for xk in self.mapGraph[xi]:
                    if xk != xj:
                        queue.append((xk, xi))
        return True

    def revise(self, xi, xj):
        revised = False
        for color in list(self.domains[xi]):
            if all(not self.isConsistent(xi, color) for color in self.domains[xi]):
                self.domains[xi].remove(color)
                revised = True
        return revised

    def backtrack(self):
        if len(self.colorsAssign) == len(self.mapGraph):
            return True
        node = self.unassign()
        for color in list(self.domains[node]):
            if self.isConsistent(node, color):
                self.colorsAssign[node] = color
                originalDomains = {n: set(d) for n, d in self.domains.items()}
                if self.ac3():
                    if self.backtrack():
                        return True
                self.colorsAssign.pop(node)
                self.domains = originalDomains
        return False

    def unassign(self):
        for node in self.mapGraph:
            if node not in self.colorsAssign:
                return node
        return None

    def solve(self):
        if self.ac3() and self.backtrack():
            self.print()
        else:
            print("No solution exists")

    def print(self):
        for node, color in self.colorsAssign.items():
            print(f"Region {node} -> Color {color}")
        print()

mapGraph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'C'],
    'C': ['A', 'B', 'D'],
    'D': ['A', 'C']
}
colors = ['Red', 'Green', 'Blue']
csp = CSPAC3(mapGraph, colors)
csp.solve()
```

# Propositional Logic Syntax and Semantic

```python
class Logic:
    def __init__(self, vars, expr):
        self.vars = vars
        self.expr = expr

    def evalExpr(self, expr, vals):
        for v, val in vals.items():
            expr = expr.replace(v, 'True' if val else 'False')
        expr = expr.replace('and', 'and').replace('or', 'or').replace('not', 'not')
        return eval(expr)

    def genComb(self):
        n = len(self.vars)
        combs = []
        for i in range(2**n):
            comb = [bool(i & (1 << j)) for j in range(n)]
            combs.append(comb)
        return combs

    def printTable(self):
        combs = self.genComb()
        header = " | ".join(self.vars) + " | " + self.expr
        print(header)
        print("-" * len(header))
        
        for comb in combs:
            vals = dict(zip(self.vars, comb))
            res = self.evalExpr(self.expr, vals)
            resStr = str(res).upper()
            print(" | ".join([str(int(val)).ljust(3) for val in comb]) + " | " + resStr)

vars = ['p', 'q']
expr = '(p and q) or (not p and not q)'
ttg = Logic(vars, expr)
ttg.printTable()
```

# Propositional Logic Inference

```python
class Logic:
    def __init__(self, stmts):
        self.stmts = stmts
    
    def modusPonens(self, p, impl):
        if p in self.stmts and impl in self.stmts:
            ant, cons = impl.split('->')
            if p == ant.strip():
                return cons.strip()
        return None

    def modusTollens(self, q, impl):
        if q in self.stmts and impl in self.stmts:
            ant, cons = impl.split('->')
            if q == f"not {cons.strip()}":
                return f"not {ant.strip()}"
        return None

    def disjunctive(self, disj, neg):
        if disj in self.stmts and neg in self.stmts:
            p, q = disj.split('or')
            if neg == f"not {p.strip()}":
                return q.strip()
            if neg == f"not {q.strip()}":
                return p.strip()
        return None

    def hypothetical(self, impl1, impl2):
        if impl1 in self.stmts and impl2 in self.stmts:
            ant1, cons1 = impl1.split('->')
            ant2, cons2 = impl2.split('->')
            if cons1.strip() == ant2.strip():
                return f"{ant1.strip()} -> {cons2.strip()}"
        return None

    def conjunction(self, p, q):
        if p in self.stmts and q in self.stmts:
            return f"{p} and {q}"
        return None

    def addition(self, p, q):
        if p in self.stmts:
            return f"{p} or {q}"
        return None


stmts = [
    "p -> q",
    "not q",
    "p",
    "p or q",
    "not p"
]

logic = Logic(stmts)
print("Modus Ponens:", logic.modusPonens("p", "p -> q"))
print("Modus Tollens:", logic.modusTollens("not q", "p -> q"))
print("Disjunctive Syllogism:", logic.disjunctive("p or q", "not p"))
print("Hypothetical Syllogism:", logic.hypothetical("p -> q", "q -> r"))
print("Conjunction:", logic.conjunction("p", "q"))
print("Addition:", logic.addition("p", "q"))
```

# Propositional Logic Resolution Theorem

```python
class Logic:
    def __init__(self, clauses):
        self.clauses = [set(clause.split()) for clause in clauses]
    
    def resolve(self, clause1, clause2):
        resolvents = set()
        for lit1 in clause1:
            for lit2 in clause2:
                if lit1 == f"not {lit2}":
                    new_clause = (clause1 - {lit1}) | (clause2 - {lit2})
                    if new_clause:
                        resolvents.add(frozenset(new_clause))
        return resolvents
    
    def prove(self, goal):
        newclauses = set(frozenset(clause) for clause in self.clauses)
        while True:
            resolvents = set()
            clauses_list = list(newclauses)
            for i in range(len(clauses_list)):
                for j in range(i + 1, len(clauses_list)):
                    resolvents |= self.resolve(clauses_list[i], clauses_list[j])
            if frozenset() in resolvents:
                return True
            if frozenset(goal.split()) in resolvents:
                return True
            if resolvents.issubset(newclauses):
                return False
            newclauses |= resolvents
    
    def printProof(self, goal):
        if self.prove(goal):
            print(f"The clause '{goal}' can be derived")
        else:
            print(f"The clause '{goal}' cannot be derived")

clauses = [
    "p or q",
    "not p or r",
    "not q or s",
    "not r or not s"
]
goal = "not p or not s"
logic = Logic(clauses)
logic.printProof(goal)
```

# Perceptron Training Rule for a single-layer perceptron

```python
import numpy as np
class Perceptron:
    def __init__(self, input_size, learning_rate=0.01, epochs=1000):
        self.weights = np.zeros(input_size + 1)
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, x):
        x_with_bias = np.insert(x, 0, 1)
        return 1 if np.dot(self.weights, x_with_bias) >= 0 else 0

    def fit(self, X, y):
        for _ in range(self.epochs):
            for xi, target in zip(X, y):
                update = self.learning_rate * (target - self.predict(xi))
                self.weights[1:] += update * xi
                self.weights[0] += update

X = np.array([[2, 3], [1, 1], [2, 1], [3, 2]])
y = np.array([1, 0, 1, 1])
perceptron = Perceptron(input_size=2)
perceptron.fit(X, y)

for xi in X:
    print(perceptron.predict(xi))
```

# Perceptron Training Rule for a muti-layer perceptron

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoidDerivative(x):
    return x * (1 - x)

class Mlp:
    def __init__(self, inputSize, hiddenSize, outputSize, learningRate=0.1, epochs=10000):
        self.inputSize = inputSize
        self.hiddenSize = hiddenSize
        self.outputSize = outputSize
        self.learningRate = learningRate
        self.epochs = epochs
        self.weightsInputHidden = np.random.uniform(size=(inputSize, hiddenSize))
        self.weightsHiddenOutput = np.random.uniform(size=(hiddenSize, outputSize))
        self.biasHidden = np.random.uniform(size=(1, hiddenSize))
        self.biasOutput = np.random.uniform(size=(1, outputSize))

    def forward(self, x):
        self.hiddenLayerInput = np.dot(x, self.weightsInputHidden) + self.biasHidden
        self.hiddenLayerOutput = sigmoid(self.hiddenLayerInput)
        self.outputLayerInput = np.dot(self.hiddenLayerOutput, self.weightsHiddenOutput) + self.biasOutput
        self.output = sigmoid(self.outputLayerInput)
        return self.output

    def backward(self, x, y, output):
        error = y - output
        outputGradient = error * sigmoidDerivative(output)
        hiddenError = outputGradient.dot(self.weightsHiddenOutput.T)
        hiddenGradient = hiddenError * sigmoidDerivative(self.hiddenLayerOutput)
        self.weightsHiddenOutput += self.hiddenLayerOutput.T.dot(outputGradient) * self.learningRate
        self.weightsInputHidden += x.T.dot(hiddenGradient) * self.learningRate
        self.biasOutput += np.sum(outputGradient, axis=0, keepdims=True) * self.learningRate
        self.biasHidden += np.sum(hiddenGradient, axis=0, keepdims=True) * self.learningRate

    def train(self, x, y):
        for _ in range(self.epochs):
            output = self.forward(x)
            self.backward(x, y, output)

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
mlp = Mlp(inputSize=2, hiddenSize=2, outputSize=1)
mlp.train(x, y)

for xi in x:
    print(mlp.forward(xi))
```

#  Neural Network Training

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
x = digits.data
y = digits.target

def trainTestSplit(x, y, testSize=0.2, randomState=None):
    if randomState is not None:
        np.random.seed(randomState)
    numSamples = x.shape[0]
    indices = np.arange(numSamples)
    np.random.shuffle(indices)
    splitIndex = int(numSamples * (1 - testSize))
    trainIndices = indices[:splitIndex]
    testIndices = indices[splitIndex:]
    xTrain = x[trainIndices]
    yTrain = y[trainIndices]
    xTest = x[testIndices]
    yTest = y[testIndices]
    return xTrain, xTest, yTrain, yTest

xTrain, xTest, yTrain, yTest = trainTestSplit(x, y, testSize=0.2, randomState=42)

mean = xTrain.mean(axis=0)
std = xTrain.std(axis=0)
xTrain = (xTrain - mean) / std
xTest = (xTest - mean) / std

inputSize = xTrain.shape[1]
hiddenSize = 32
outputSize = 10
learningRate = 0.01
epochs = 1000

np.random.seed(42)
w1 = np.random.randn(inputSize, hiddenSize) * 0.01
b1 = np.zeros((1, hiddenSize))
w2 = np.random.randn(hiddenSize, outputSize) * 0.01
b2 = np.zeros((1, outputSize))

def relu(z):
    return np.maximum(0, z)

def reluDerivative(z):
    return (z > 0).astype(float)

def softmax(z):
    expZ = np.exp(z - np.max(z, axis=1, keepdims=True))
    return expZ / expZ.sum(axis=1, keepdims=True)

def crossEntropyLoss(yTrue, yPred):
    return -np.mean(np.log(yPred[np.arange(len(yTrue)), yTrue]))

for epoch in range(epochs):
    z1 = xTrain.dot(w1) + b1
    a1 = relu(z1)
    z2 = a1.dot(w2) + b2
    a2 = softmax(z2)
    loss = crossEntropyLoss(yTrain, a2)

    dz2 = a2
    dz2[np.arange(len(yTrain)), yTrain] -= 1
    dz2 /= len(yTrain)

    dw2 = a1.T.dot(dz2)
    db2 = np.sum(dz2, axis=0, keepdims=True)

    da1 = dz2.dot(w2.T)
    dz1 = da1 * reluDerivative(z1)

    dw1 = xTrain.T.dot(dz1)
    db1 = np.sum(dz1, axis=0, keepdims=True)

    w1 -= learningRate * dw1
    b1 -= learningRate * db1
    w2 -= learningRate * dw2
    b2 -= learningRate * db2

    if epoch % 100 == 0:
        print(f'epoch {epoch}, loss: {loss}')

z1Test = xTest.dot(w1) + b1
a1Test = relu(z1Test)
z2Test = a1Test.dot(w2) + b2
a2Test = softmax(z2Test)

yPred = np.argmax(a2Test, axis=1)

accuracy = np.mean(yPred == yTest)
print(f'test accuracy: {accuracy * 100:.2f}%')

plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(xTest[i].reshape(8, 8), cmap='gray')
    plt.axis('off')
plt.show()
```
